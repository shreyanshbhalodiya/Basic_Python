```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,3,5,7]
y = [2,4,6,8]
plt.title("Test Plot")
plt.xlabel(" X axis")
plt.ylabel(" Y axis")
plt.plot(x ,y)
plt.show()
```

```python
x = [1,3,5,7]
y = [2,4,6,8]
z = [10,8,6,5]
plt.title("Test Plot")
plt.xlabel(" X axis")
plt.ylabel(" Y axis & Z axis")
plt.plot(x ,y)
plt.plot(x , z)
plt.legend(["This is y "," This is z"])
plt.show()
```
