```python
def add(a , b, *args):
    print(a , b , args)
    total = 0
    for i in args:
        total += i
    return total

print(add(2,45,45,78,12,12,41,14,14))
```

```python
def add(a , b, *args , **kwargs):
    print(a , b , args , kwargs)
    total = 0
    for i in args:
        total += i
    return total

print(add(2,45,45,78,12,12,41,14,14))
```

```python
def add(a , b, *args , **kwargs):
    print(kwargs)
    total = 0
    for i in args:
        total += i
    return total

print(add(2,45,45,78,12,12,41,14,14, operation = "minus"))
```
